```python
all_movies = pd.DataFrame()
```

```python
for i in range(1,500):
  requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=17bd2d67dc3c42c0db9bf2a0c0d85dc6&language=en-US&page={}'.format(i))
  temp_df = pd.DataFrame(response.json()['results'])[['title', 'overview', 'release_date', 'genre_ids', 'vote_average']]
  all_movies = all_movies.append(temp_df, ignore_index=True)
```

```python
all_movies
```

```python
# get genres details
response1 = requests.get('https://api.themoviedb.org/3/genre/movie/list?api_key=17bd2d67dc3c42c0db9bf2a0c0d85dc6&language=en-US')
```

```python
response1.json()['genres']
```

```python
genres = dict()
for i in response1.json()['genres']:
  genres[i['id']] = i['name']
genres
```

```python
all_movies['genres'] = all_movies['genre_ids'].apply(lambda x: [genres[i] for i in x])
```
